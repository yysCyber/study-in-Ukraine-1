# 3. Build NN

### 3.1. Data Type Conversion

```python
train_x_data_cast = tf.cast(train_x_data, tf.float32)
test_x_data_cast = tf.cast(test_x_data, tf.float32)
```

```python
train_x_data_cast
```

```python
test_x_data_cast
```

### 3.2. Define NN Parameters

```python
train_batches = tf.data.Dataset.from_tensor_slices((train_x_data_cast, train_y_data)).batch(32)
```

```python
test_batches = tf.data.Dataset.from_tensor_slices((test_x_data_cast, test_y_data)).batch(32)
```

```python
# Random
w = tf.Variable(tf.random.truncated_normal([4, 3], stddev=0.1))
```

```python
# Random
b = tf.Variable(tf.random.truncated_normal([3], stddev=0.1))
```

```python
learnin_rate = 0.1
```

```python
train_rounds = 500
```

```python
predicted_result = []
```

### 3.3. Training

```python
for train_round in range(train_rounds):
    for step, (x1, y1) in enumerate(train_batches):
        with tf.GradientTape() as tape:
            y = tf.matmul(x1, w) + b
            y = tf.nn.softmax(y)
            y_correct = tf.one_hot(y1, depth=3)
            loss = tf.reduce_mean(tf.square(y_correct - y))
        grads = tape.gradient(loss, [w, b])
        w.assign_sub(learnin_rate * grads[0])
        b.assign_sub(learnin_rate * grads[1])
```

### 3.4. Testing

```python
for x2, y2 in test_batches:
    y = tf.matmul(x2, w) + b
    y = tf.nn.softmax(y)
    predicted_result.append(tf.argmax(y, axis=1))
```
